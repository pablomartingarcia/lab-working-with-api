```python
import requests
import pandas as pd

headers = {
    'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
    'x-rapidapi-key': "9613ca6ca5msh62659872159916fp1a0a2cjsndd9a3e6ff324"
}


def get_airport_codes(location):
    url = "https://aerodatabox.p.rapidapi.com/airports/search/term"
    querystring = {"q": location.lower(), "limit": "10"}

    try:
        response = requests.get(url, headers=headers, params=querystring)
        response.raise_for_status()  
        data = response.json()['items'] if 'items' in response.json() else location
    except requests.exceptions.RequestException as e:
        print("Error fetching airport data:", e)
        return None

    return data


def choose_airport_code(location):
    codes = get_airport_codes(location)

    while True:
        if not codes:
            print('Try another location. Unable to find code.')
            location = input('Insert location:')
            codes = get_airport_codes(location)
        elif len(set([code['shortName'] for code in codes])) != 1:
            print('More than one location found. Insert one of the following airports:')
            for code in codes:
                print(code['shortName'])
            location = input('Insert location:')
            codes = get_airport_codes(location)
        else:
            return codes[0]['icao']


def get_flight_distance_time(from_, to_):
    url = f"https://aerodatabox.p.rapidapi.com/airports/icao/{from_}/distance-time/{to_}"

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  
        data = pd.json_normalize(response.json()).T
    except requests.exceptions.RequestException as e:
        print("Error fetching flight distance and time:", e)
        return None

    return data


location = input('Insert location:')
airport_code = choose_airport_code(location)
print("Chosen airport code:", airport_code)

from_ = input('Enter departure airport code:')
to_ = input('Enter destination airport code:')
flight_data = get_flight_distance_time(from_, to_)

if flight_data is not None:
    print("Flight distance and time data:")
    print(flight_data)
```
